Ex 8: L1 and L2 should be identical. Write a program that finds the missing number in L2.

```python
from random import randint
L1 = list(range(1,101))
L2 = L1.copy()
L2.remove(randint(1,101))
missing_numbers=[]
for num in L1:
  if num not in L2:
    missing_numbers.append(num)

print("List 1: ",L1)
print("List 2: ",L2)
print("Missing number list in L2: ",missing_numbers)
```

```python
# Test with another way
from random import randint
L1 = list(range(1,101))
L2 = L1.copy()
L2.remove(randint(1,101))

# Since only one integer is remove from list 2, we can using addition to find the only one missing number.
missing_numbers=sum(L1)-sum(L2)

print("List 1: ",L1)
print("List 2: ",L2)
print("Missing number is: ",missing_numbers)
```

Ex 9: Write a program to copy the ls1 = ["apple", "banana", "cherry"] to a new list. Add the element "blueberry" to the new list, but keep the original list the same.

```python
ls1 = ["apple", "banana", "cherry"]
print("Old list is: ", ls1)
new_lst=ls1.copy()
print("After copied ls1, New list is: ", new_lst)
new_lst=new_lst+["blueberry"]
print("After added a new element to new list: ",new_lst)
print("Old list remain as original: ",ls1)
```

# Ex 4: Daily Maths

You are working in a retail shop. Write a program that computes the total cost for a customer that bought 3 differnt types of fruits : apple, orange, mango. Ask the user to enter the quantity and price for each type.

```python
apple_qty=int(input("Enter the quantity of apple: "))
apple_price=float(input("Enter the price per apple: "))
apple_total=apple_qty*apple_price

orange_qty=int(input("Enter the quantity of orange: "))
orange_price=float(input("Enter the price per orange: "))
orange_total=orange_qty*orange_price

mango_qty=int(input("Enter the quantity of mango: "))
mango_price=float(input("Enter the price per mango: "))
mango_total=mango_qty*mango_price

total_cost=apple_total+orange_total+mango_total

print("The total cost for your buying is: ", total_cost)
```

# Ex 6: Random Number
Write a program that generates a random number, x, between 1 and 50, a random number y between 2 and 5, and computes x^y.

```python
from random import randint
x = randint(1,50)
y = randint(2,5)
print('x= A random number between 1 and 50: ', x)
print('y= A random number between 2 and 5: ', y)
print('x to the power of y is: ', x**y)
```

# Ex 9: Computing

Write a program that asks the user for a number of seconds and prints out how many minutes and seconds that is. For instance, 200 seconds is 3 minutes and 20 seconds.

```python
seconds=int(input("Enter a number of seconds: "))
minutes=seconds//60
remain_seconds=seconds%60
print("{} minutes and {} seconds".format(minutes,remain_seconds))
```

# Ex 10: Using Math Module

Write a program that asks the user for a number and then prints out the sine, cosine, and tangent of that number.

```python
import math
num = float(input("Enter a number:"))
sine_value = math.sin(num)
print("Sine value of the given number is: ",sine_value)
cosine_value = math.cos(num)
print("Cosine value of the given number is: ",cosine_value)
tangent_value = math.tan(num)
print("Tangent value of the given number is: ",tangent_value)
```
